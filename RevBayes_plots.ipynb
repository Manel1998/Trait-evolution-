```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('./output/simple_BM.log' , sep='\t')
print(df)
```

```python
sigma_values = df['sigma2']

```

```python
plt.figure(figsize=(10, 6))
sns.histplot(sigma_values, bins=30, kde=True)
plt.title('Posterior Distribution of Sigma')
plt.xlabel('Sigma')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(sigma_values)
plt.title('Trace Plot of Sigma')
plt.xlabel('Iteration')
plt.ylabel('Sigma')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x=sigma_values)
plt.title('Boxplot of Sigma Values')
plt.xlabel('Sigma')
plt.show()

```

```python
plt.figure(figsize=(12, 6))
sns.histplot(sigma_values, bins=30, kde=True)
plt.xlim([min(sigma_values), max(sigma_values)])  # Adjust limits for better zoom
plt.title('Density Plot of Sigma Values')
plt.xlabel('Sigma')
plt.ylabel('Density')
plt.show()

```

```python
plt.figure(figsize=(12, 6))
plt.scatter(df.index, sigma_values, alpha=0.5, edgecolors='w', s=50)
plt.title('Scatterplot of Sigma Values Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Sigma')
plt.grid(True)
plt.show()

```
