```python
import pandas as pd

# Load df1 once
df1 = pd.read_csv('genes_GA', sep='\t')
df1 = df1.iloc[1:, 1:]  # Remove first row and first column



GA = ['GA.iso.E', 'GA.iso.O', 'GA.iso.T', 'GA.iso.Te', 'GA.iso.br'] 

for file in GA:
    # Read each file as df
    df = pd.read_csv(file, sep='\t')
    print(f"\nProcessing {file}")
    print(df)

    # Merge df1 with df
    df2 = pd.merge(df1, df, left_index=True, right_index=True)
    df2.columns.values[0] = ''
    # Print or save df2
    print(f"\nMerged DataFrame for {file}:")
    print(df2)


    prefix = file.split('.')[0]  # Get the part before the extension (e.g., 'GA.iso')
    suffix = file.split('.')[-1]  # Get the last part after the last dot (e.g., 'E', 'O', 'T', 'Te')
    output_file = f"data_{prefix}_{suffix}.csv"  
    df2.to_csv(output_file, sep='\t', index=False)
    print(f"Merged DataFrame saved to {output_file}")
    
```

```python
#RNA_SEQ_ANALYSIS 
#CREATING A GENERAL COLDATA TABELE FOR GA 

df3 = pd.read_csv('GA.isoform.counts.matrix', sep = '\t')
df_transposed = df3.transpose()
index = list(df_transposed.index)
print(index)

```

```python
print(index)
```

```python
coldata = pd.DataFrame()
coldata['ID'] = index
coldata = coldata.iloc[1:, :]

print(coldata)
```

```python
# Define the mapping logic for sex and species based on the ID names
def get_sex(id_name):
    if '_Fbr_' in id_name or '_O_' in id_name:
        return 'Female'
    elif '_E_' in id_name or '_T_' in id_name:
        return 'Undefined'
    else:
        return 'Male'

def get_species(id_name):
    if '_Ga_' in id_name:
        return 'Goodea atripinnis'
    elif '_Gm_' in id_name:
        return 'Girardinichthys multiradiatus'
    elif '_Xc_' in id_name:
        return 'Xenoophorus captivus'
    elif '_Xr_' in id_name:
        return 'Xenotaenia resolanae'
    else:
        return 'Unknown species'

def get_samples(id_name): 
    if '_Br_' in id_name or '_Fbr_' in id_name:
        return 'Brain'
    elif '_O_' in id_name:
        return 'Ovaries'
    elif '_E_' in id_name:
        return 'Embryo'
    elif '_T_' in id_name:
        return 'Trophotaenia'
    elif '_Te_' in id_name:
        return 'Testes'

# Apply the functions to the "ID" column to get the "sex" and "species" columns
coldata['sex'] = coldata['ID'].apply(get_sex)
coldata['species'] = coldata['ID'].apply(get_species)
coldata['sample']= coldata['ID'].apply(get_samples)

# Print the resulting DataFrame
print(coldata)
```

```python
coldata.to_csv('coldata.csv', index=False)
```
